## Helper Functions

```python
def get_chess_info(game_number):
    # Create URL
    url = 'https://www.chess.com/live/game/{}'.format(str(game_number))
    
    try:
        # Grab it, Soup it, Find what we want
        text = requests.get(url).text
        b = BeautifulSoup(text, 'html.parser')
        game_result = b.find('meta', {'name':'description'})['content']

        if text[:5] == 'Login':
            print('none!')
            return None
        else: 
            return game_result
    except:
        print('error getting info')
        return None
```

```python
def parse_text(text):
    try:
        important = text.split('. The average chess')[0]
        container1 = important.split('. ')[0]
        container2 = important.split('. ')[1]\

        player1 = container1.split('vs ')[0].split(' ')[0]
        player1_elo = int(container1.split('vs ')[0].split(' ')[1].replace('(', '').replace(')', ''))

        player2 = container1.split('vs ')[1].split(' ')[0]
        player2_elo = int(container1.split('vs ')[1].split(' ')[1].replace('(', '').replace(')', ''))

        winner = container2.split(' ')[0]

        print(player1, player1_elo, player2, player2_elo, winner)
    
        return(player1, player1_elo, player2, player2_elo, winner)
    except: 
        print('error parsing text')
        return(None, None, None, None, None)
    
```

```python
count = 0 

with open('chess_data.csv', 'a') as csvfile:
    
    csvfile.write("\n")
    fieldnames = ['player1', 
                  'player1_elo', 
                  'player2', 
                  'player2_elo', 
                  'winner',
                  'game_id']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    while count <= 200:
        i = randrange(10000, 4107999562)

        time.sleep(.5)
        text = get_chess_info(i)

        print(i, count)

        player1, player1_elo, player2, player2_elo, winner = parse_text(text)
        
        if player1:
            writer.writerow({'player1': player1, 
                             'player1_elo': player1_elo,
                             'player2': player2,
                             'player2_elo': player2_elo,
                             'winner': winner, 
                             'game_id': i
                            })
        
        count += 1

```
